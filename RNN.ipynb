## Sentiment Analysis

```python
from keras.datasets import imdb
from keras import Sequential
from keras.layers import Dense, SimpleRNN, Embedding, Flatten
```

```python
(x_train, y_train),(x_test, y_test)= imdb.load_data()
```

```python
# y_test
```

```python
print(len(x_train[0]))
print(len(x_train[1]))
```

```python
# Different data point in the dataset are in different sequence. to make it in same in sequence.
x_train=pad_sequences(x_train, padding='post', maxlen=50)
x_test=pad_sequences(x_test, padding='post', maxlen=50)
```

```python
x_train[0]
```

```python
model=Sequential()
model.add(SimpleRNN(32, input_shape=(50,1), return_sequences=False))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
```

# Applying Embedding

```python
doc=[
    'recurrent neural network',
    'artificial network',
    'connection between nodes',
    'can create a cycle',
    'allowing output',
    'som nodes to affect subsequent',
    'exhibit temporal',
    'dynamic bahavior',
    'type of neural network',
    'affect subsequent'
]
```

```python
from keras.preprocessing.text import Tokenizer
tokenizer=Tokenizer(oov_token='<nothing>')
```

```python
tokenizer.fit_on_texts(doc)
```

```python
tokenizer.word_index
```

```python
len(tokenizer.word_index)
```

```python
sequences=tokenizer.texts_to_sequences(doc)
```

```python
from keras.utils import pad_sequences
sequence=pad_sequences(sequences, padding='post')
```

```python
sequence
```

```python
model=Sequential()
model.add(Embedding(24, output_dim=2, input_length=5)) # output_dim=2 means i want to represent each word in to 2 dimensions vector.
                                                      # 23 means we have 24 unique word in corpus.
                                                      # inputlength = 5 have input shape of five.
```

```python
model.summary()
```

```python
model.compile('adam','accuracy')
pred=model.predict(sequence)
```

## Imdb using Embedding

```python
from keras.datasets import imdb
from keras.preprocessing.text import Tokenizer
from keras.utils import pad_sequences
from keras import Sequential
from keras.layers import Dense, SimpleRNN, Embedding, Flatten
```

```python
(x_train, y_train),(x_test, y_test)=imdb.load_data()
```

```python
x_train=pad_sequences(x_train, padding='post', maxlen=50)
x_test=pad_sequences(x_test, padding='post', maxlen=50)
```

```python
x_train.shape
```

```python
model=Sequential()
model.add(Embedding(10000, output_dim=2, input_length=50))
model.add(SimpleRNN(32,input_shape=(50,1), return_sequences=False))
model.add(Dense(1, activation='sigmoid'))
model.summary()
```

```python
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['acc'])
histoty=model.fit(x_train, y_train, epochs=5, validation_data=(x_test, y_test))
```
